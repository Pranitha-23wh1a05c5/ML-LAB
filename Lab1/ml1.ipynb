```python
import math
import statistics
import numpy as np
from scipy import stats

# Input arrays
a = np.array(list(map(float, input("Enter elements of array A: ").split())))
b = np.array(list(map(float, input("Enter elements of array B: ").split())))

print("Addition:", np.add(a, b))
print("Subtraction:", np.subtract(a, b))
print("Multiplication:", np.multiply(a, b))
print("Division:", np.divide(a, b))
print("Exponentiation:", np.power(a, b))
print("Modulus:", np.mod(a, b))

print("Mean of A:", statistics.mean(a))
print("Mean of B:", statistics.mean(b))

print("Standard Deviation of A:", statistics.stdev(a))
print("Standard Deviation of B:", statistics.stdev(b))

print("Square root of A:", np.sqrt(a))
print("Factorial of first element of A:", math.factorial(int(a[0])))

print("Z-score of A:", stats.zscore(a))
print("Z-score of B:", stats.zscore(b))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load iris dataset
iris = load_iris()

# Convert to Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add target column
df['species'] = iris.target
```

```python
df.head()
```

```python
df.info()
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Replace numeric species with names
df['species'] = df['species'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})
```

```python
#Filter data
setosa_df = df[df['species'] == 'Setosa']
setosa_df.head()
```

```python
#Matplotlib Library Functions (Data Visualization)
#Line Plot
plt.plot(df['sepal length (cm)'])
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
#Scatter Plot
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
#Histogram
plt.hist(df['petal width (cm)'], bins=10)
plt.title("Petal Width Distribution")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar Chart(Species Count)
df['species'].value_counts().plot(kind='bar')
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
#FIND-S Algorithm Implementation

data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

# Initialize hypothesis with first positive example
hypothesis = None
for row in data:
    if row[-1] == 'Yes':
        hypothesis = row[:-1]
        break

# Apply FIND-S
for row in data:
    if row[-1] == 'Yes':
        for i in range(len(hypothesis)):
            if hypothesis[i] != row[i]:
                hypothesis[i] = '?'

print("Most Specific Hypothesis:")
print(hypothesis)

```
